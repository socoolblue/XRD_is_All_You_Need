```python
drop_rate= 0.1
drop_rate_2= 0.1
initializer = keras.initializers.RandomNormal()

inputs = keras.Input(shape=(8192, 1))

c1=layers.Conv1D(filters=16, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(), 
                        kernel_initializer=initializer)(inputs)
p1=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c1)
p1=layers.Dropout(rate=drop_rate, seed=seed_num)(p1)
c2=layers.Conv1D(filters=16, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(), 
                        kernel_initializer=initializer)(p1)
p2=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c2)
p2=layers.Dropout(rate=drop_rate, seed=seed_num)(p2)
c3=layers.Conv1D(filters=32, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(),
                        kernel_initializer=initializer)(p2)
p3=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c3)
p3=layers.Dropout(rate=drop_rate, seed=seed_num)(p3)
c4=layers.Conv1D(filters=32, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(),
                        kernel_initializer=initializer)(p3)
p4=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c4)
p4=layers.Dropout(rate=drop_rate, seed=seed_num)(p4)
c5=layers.Conv1D(filters=64, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(), 
                        kernel_initializer=initializer)(p4)
p5=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c5)
p5=layers.Dropout(rate=drop_rate, seed=seed_num)(p5)
c6=layers.Conv1D(filters=64, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(), 
                        kernel_initializer=initializer)(p5)
p6=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c6)
p6=layers.Dropout(rate=drop_rate, seed=seed_num)(p6)
c7=layers.Conv1D(filters=64, kernel_size=3, strides=1,padding='same',activation=tf.keras.layers.LeakyReLU(),
                         kernel_initializer=initializer)(p6)
p7=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c7)
p7=layers.Dropout(rate=drop_rate, seed=seed_num)(p7)
c8=layers.Conv1D(filters=128, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(), 
                        kernel_initializer=initializer)(p7)
p8=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c8)
p8=layers.Dropout(rate=drop_rate, seed=seed_num)(p8)
c9=layers.Conv1D(filters=256, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(),
                        kernel_initializer=initializer)(p8)
p9=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c9)
p9=layers.Dropout(rate=drop_rate, seed=seed_num)(p9)
c10=layers.Conv1D(filters=256, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(), 
                        kernel_initializer=initializer)(p9)
p10=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c10)
p10=layers.Dropout(rate=drop_rate, seed=seed_num)(p10)
c11=layers.Conv1D(filters=512, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(), 
                         kernel_initializer=initializer)(p10)
p11=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c11)
p11=layers.Dropout(rate=drop_rate, seed=seed_num)(p11)
c12=layers.Conv1D(filters=256, kernel_size=3, strides=1,padding='same', activation=tf.keras.layers.LeakyReLU(), 
                         kernel_initializer=initializer)(p11)
p12=layers.MaxPooling1D(pool_size=2, strides=2, padding='same')(c12)
f1 = layers.Flatten()(p12)


inputs_2 = keras.Input(shape=(1, 100))
d1_2 = layers.Dense(128, kernel_initializer=initializer, activation=tf.keras.layers.LeakyReLU())(inputs_2)
d1_2 = layers.Dropout(rate=drop_rate, seed=seed_num)(d1_2)
d2_2 = layers.Dense(128, kernel_initializer=initializer, activation=tf.keras.layers.LeakyReLU())(d1_2)
d2_2 = layers.Dropout(rate=drop_rate, seed=seed_num)(d2_2)
d3_2 = layers.Dense(64, kernel_initializer=initializer, activation=tf.keras.layers.LeakyReLU())(d2_2)
d3_2 = layers.Dropout(rate=drop_rate, seed=seed_num)(d3_2)
d4_2 = layers.Dense(64, kernel_initializer=initializer, activation=tf.keras.layers.LeakyReLU())(d3_2)
d4_2 = layers.Dropout(rate=drop_rate, seed=seed_num)(d4_2)
d5_2 = layers.Dense(32, kernel_initializer=initializer, activation=tf.keras.layers.LeakyReLU())(d4_2)
d5_2 = layers.Dropout(rate=drop_rate, seed=seed_num)(d5_2)
d6_2 = layers.Dense(32, kernel_initializer=initializer, activation=tf.keras.layers.LeakyReLU())(d5_2)
d7_2 = layers.Dropout(rate=drop_rate, seed=seed_num)(d6_2)
d8_2 = layers.Dense(16, kernel_initializer=initializer, activation=tf.keras.layers.LeakyReLU())(d7_2)
f2= layers.Flatten()(d8_2)


concatenated = concatenate([f1, f2])
dense_all_2 = layers.Dense(128, kernel_initializer=initializer, activation=tf.keras.layers.LeakyReLU())(concatenated)
dense_all_2 = layers.Dropout(rate=drop_rate, seed=seed_num)(dense_all_2)
dense_all_3 = layers.Dense(16, kernel_initializer=initializer, activation=tf.keras.layers.LeakyReLU())(dense_all_2)
dense_all_3 = layers.Dropout(rate=drop_rate, seed=seed_num)(dense_all_3)
outputs = layers.Dense(1, kernel_initializer=initializer)(dense_all_3)
model = tf.keras.Model(inputs=[inputs, inputs_2], outputs=[outputs])
```

```python
h5= h5py.File('mp_encoding.h5', 'r')
stack=h5['data'][:,:]

np.random.shuffle(stack)
ratio_ = int(stack.shape[0]*0.2)


#minmax norm
#min_, max_ = np.min(stack[:,-2]), np.max(stack[:,-2])
#stack[:,-2] = (stack[:,-2] - min_)/(max_ - min_)

x_data_tr_1 = stack[:-ratio_,:-107]/100
x_data_tr_2 = stack[:-ratio_,-107:-7]
y_data_tr = (stack[:-ratio_,-2]).reshape(-1,1)

x_data_te_1 = stack[-ratio_:,:-107]/100
x_data_te_2 = stack[-ratio_:,-107:-7]
y_data_te = (stack[-ratio_:,-2]).reshape(-1,1)


x_data_tr_1 = x_data_tr_1.r'eshape(-1,x_data_tr_1.shape[1],1)
x_data_tr_2 = x_data_tr_2.reshape(-1,1,x_data_tr_2.shape[1])
x_data_te_1 = x_data_te_1.reshape(-1,x_data_te_1.shape[1],1)
x_data_te_2 = x_data_te_2.reshape(-1,1,x_data_te_2.shape[1])


print(stack.shape)
print(x_data_tr_1.shape, x_data_te_1.shape)
print(x_data_tr_2.shape, x_data_te_2.shape)
print(y_data_tr.shape, y_data_te.shape)
```

```python
# 4-fold
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import KFold
num=0

for tr, te in KFold(n_splits=4, shuffle=True).split(x_data_tr_1, y_data_tr):
    num+=1
    print("========================================================================", num,"th fold reslt")
    opt = tf.keras.optimizers.Adam(learning_rate=0.0002)
    model.compile(optimizer=opt, loss = tf.keras.losses.MeanAbsoluteError())
    
    checkpoint_path = "Kfold_form_e_new/%s_training_{epoch:04d}.ckpt" %num
    cp_callback = tf.keras.callbacks.ModelCheckpoint(checkpoint_path, save_weights_only=True, verbose=1, save_freq=16300)
    
    model.fit([x_data_tr_1[tr], x_data_tr_2[tr]], y_data_tr[tr], epochs=100, shuffle=True, batch_size=1024, callbacks = [cp_callback],
                       validation_data=([x_data_te_1,x_data_te_2],y_data_te))
    
    print("========================================================================", num,"th fold validation accuracy")
    model.evaluate([x_data_tr_1[te], x_data_tr_2[te]], y_data_tr[te])
    print("========================================================================", num,"th fold hold-out test accuracy")
    model.evaluate([x_data_te_1, x_data_te_2], y_data_te, batch_size=2048)
    fig_y = model.predict([x_data_te_1,x_data_te_2])

    plt.figure(figsize=(8, 8))
    plt.scatter(y_data_te, fig_y, s=8)

    plt.plot([-8, 8], [-8, 8])

    plt.xlim(-8, 8) 
    plt.ylim(-8, 8)
    plt.show()

    
```
